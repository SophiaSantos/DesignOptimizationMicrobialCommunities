```python
from reframed import load_cbmodel
from cobra.io import read_sbml_model
from mewpy.simulation import get_simulator
from mewpy.simulation.environment import Environment
```

```python
model = load_cbmodel('community_eco_sce.xml')
```

```python
medium = dict()
Environment.empty(model, inplace=True)

minimal_medium_es = ['R_EX_M_xyl__D_e_pool', 'R_EX_ac_e_iMM904', 'R_EX_M_mobd_e_pool', 'R_EX_M_cl_e_pool', 'R_EX_M_cobalt2_e_pool',
                     'R_EX_M_mn2_e_pool', 'R_EX_M_mg2_e_pool', 'R_EX_M_ca2_e_pool', 'R_EX_M_k_e_pool', 
                     'R_EX_M_zn2_e_pool', 'R_EX_xyl__D_e_iAF1260', 'R_EX_M_so4_e_pool', 
                     'R_EX_M_cu_e_pool', 'R_EX_M_fe3_e_pool', 'R_EX_M_o2_e_pool', 'R_EX_M_pi_e_pool', 
                     'R_EX_M_nh4_e_pool']

for exchange in minimal_medium_es:
            if exchange != 'R_EX_M_xyl__D_e_pool' and exchange != 'R_EX_ac_e_iMM904':
                medium[exchange]=(-1000, 0)
            else:
                medium[exchange]=(-10, 1000)

# Somente existe tyrosina produzida pelos organismos
medium['R_EX_M_tyr__L_e_pool']=0
```

```python
sim = get_simulator(model, envcond=medium)
```

## Optimização

```python
from mewpy.optimization import EA
from mewpy.optimization.evaluation import CandidateSize, MolecularWeight, TargetFlux, BPCY
from mewpy.problems import RKOProblem, ROUProblem
```

```python
BIOMASS = list(sim.objective.keys())[0]
BIOMASS
```

```python
TARGET = 'R_EX_tyr__L_e_iAF1260'
```

```python
f1 = BPCY(BIOMASS,TARGET)
f2 = TargetFlux(TARGET)
```

```python
essential = sim.essential_reactions()
```

```python
KO_targets = []
for rxn in sim.reactions:
    if rxn != 'R_ATPM_iAF1260':
        if rxn.endswith('iAF1260') and rxn not in essential:
            KO_targets.append(rxn)
            
print(len(KO_targets))
```

```python
for rxn in KO_targets:
    if rxn.startswith('R_EX_'):
        KO_targets.remove(rxn)
    if rxn.startswith('R_ATPS'):
        KO_targets.remove(rxn)
    if rxn.endswith('tex_iAF1260'):
        KO_targets.remove(rxn)
    if rxn.endswith('pp_iAF1260'):
        KO_targets.remove(rxn)

print(len(KO_targets))
```

```python
problem = RKOProblem(model, fevaluation=[f1,f2], target=KO_targets, candidate_max_size=5)
ea = EA(problem, max_generations=100)
```

```python
ea.run(simplify=False)
```

```python
problem = RKOProblem(model, fevaluation=[f1,f2], target=KO_targets, candidate_max_size=3)
ea = EA(problem, max_generations=100)
```
